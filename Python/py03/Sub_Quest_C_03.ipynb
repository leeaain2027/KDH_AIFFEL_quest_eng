```python
import math as m

def best_calc(prompt):
    while True:
        try:
            return int(input(prompt))
        except ValueError:
            print("잘못된 입력입니다. 정수를 입력해주세요.")

while True:
    # 정수 입력
    num1 = best_calc("첫 번째 정수를 입력하세요: ")
    num2 = best_calc("두 번째 정수를 입력하세요: ")

    # 연산자 입력
    yeon = input("연산자를 입력하세요 (+, -, *, /, **): ")

    try:
        if yeon == "+":
            result = num1 + num2
        elif yeon == "-":
            result = num1 - num2
        elif yeon == "*":
            result = num1 * num2
        elif yeon == "/":
            result = num1 / num2
        elif yeon == "**":
            result = m.pow(num1, num2)
        else:
            print("지원하지 않는 연산자입니다.")
            continue

        print("결과:", result)

    except ZeroDivisionError:
        print("0으로 나눌 수 없습니다.")

            # 계속 여부 확인
    cont = input("계속 계산하시겠습니까? (y/n): ")
    if cont.lower() != "y":
        print("계산기를 종료합니다.")
        break
```
